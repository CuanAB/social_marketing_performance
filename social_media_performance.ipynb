## Cleaning data
---

### check missing

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

Note: CTR và Clicks chỉ được ghi nhận cho một tập con các bài đăng, cho thấy không phải tất cả nội dung đều hướng đến mục tiêu chuyển đổi.
- Chỉ những bài post có gắn link / CTA mới track click
- Các bài giải trí, meme, video ngắn → không đo click

```python
# fill missing values = 0
df['Clicks'] = df['Clicks'].fillna(0)
df['Click_Through_Rate'] = df['Click_Through_Rate'].fillna(0)
df
```

```python
df.isnull().sum()
```

=> Chỉ có số liệu bị thiếu được tìm thấy ở mục Số lần nhấp chuột và Tỷ lệ nhấp chuột. Những giá trị bị thiếu này được hiểu là các bài đăng không có liên kết có thể nhấp chuột và do đó được thay thế bằng số 0 để đảm bảo việc tổng hợp và phân tích chính xác.

```python
# check duplicates
df.duplicated().sum()
```

### Convert data

```python
df.dtypes
```

### Tạo, tách cột thời gian

```python
# Tạo các cột thời gian quan trong
df['Post_Hour'] = df['Post_Published_At'].dt.hour # Tạo lại Post_Hour từ Post_Published_At để đảm bảo chính xác
df['Day_Name'] = df['Post_Published_At'].dt.day_name()
df['Month'] = df['Post_Published_At'].dt.month
```

### Tạo KPI phụ

```python
# tạo cột KPI phụ
df['Engagement_per_View'] = df['Engagement'] / df['Views']
df['Video_View_Rate'] = df['Video_Views'] / df['Views']
```

=> Thêm các chỉ số  như mức độ tương tác trên mỗi lượt xem và tỷ lệ xem video đã được tạo ra để đánh giá hiệu quả nội dung tốt hơn
